#### ***Data Analysis*** ####

##### *****1) Data Quality check***** #####

```python
#Convert date columns into datetime
df['Order Date'] = pd.to_datetime(df['Order Date'],errors = 'coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

#Check converstion
print("\nDate columns converted\n")
print(f"Order Date: {df['Order Date'].dtype}")
print(f"Ship Date: {df['Ship Date'].dtype}")
```

#### ***Univariate Analysis*** ####

```python
#univeriate analysis is the simplest form of statistical analysis. 
#As the name of suggests-"Uni" means "One" and "variate" means "Variable"- it is analysis of single variable

print("\n****** Categorical variable ******\n")

categorical_col = ['Customer Segment','Category','Region','Payment Mode','Delivery Status','Auto Reorder']

for col in categorical_col:
    print(f"----{col}----\n")
    print(df[col].value_counts())

    #visualization
    plt.figure(figsize=(6,6))
    df[col].value_counts().plot(kind="bar",color="skyblue")
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
```

#### ***Numerical variables Analysis*** ####

```python
df.columns
```

```python
print("****** Numerical Variables ******")

numerical_col = ['Quantity','Unit Price','Discount (%)','Sales Amount','Cost Price','Profit','Stock left','Reorder Quantity']

for col in numerical_col:
    print(f"\n------ {col} -------\n")
    print(f"Mean : {df[col].mean():.2f}")
    print(f"Median : {df[col].median():.2f}")
    print(f"Std dev : {df[col].std():.2f}")
    print(f"Max : {df[col].max():.2f}")
    print(f"Min : {df[col].min():.2f}")

    #visualization
    plt.figure(figsize=(8,4))

    plt.subplot(1,2,1)
    df[col].hist(bins=30, color='lightgreen', alpha=0.7)
    plt.title(f"Distibution of {col}")

    plt.subplot(1,2,2)
    df.boxplot(column=col)
    plt.title(f"Boxplot of {col}")

    plt.tight_layout()
    plt.show()  
```

#### ***Bivariate Analysis*** ####

```python
#Bivariate analysis is the statistical analysis of two variables at the same time. The core performance is investigate between them

print("\n***** Sales & Profit by category ******\n")

category_performance = df.groupby('Category').agg({
    "Sales Amount" : "sum",
    "Profit" : "sum",
    "Order ID": "count"
}).rename(columns={'Order ID':'Order Count'})

category_performance['Profit Margin'] = (category_performance['Profit'] / (category_performance['Sales Amount']) * 100)
print(category_performance,"\n\n")

#visualization
fig, axes = plt.subplots(2, 2, figsize=(14,10))

# Sales by Category
category_performance['Sales Amount'].plot(kind='bar', ax=axes[0,0], color='blue', alpha=0.7)
axes[0,0].set_title('Total Sales by Category')
axes[0,0].set_xlabel('Category')
axes[0,0].set_ylabel('Sales Amount')
axes[0,0].tick_params(axis='x', rotation=30)

# Profit by Category
category_performance['Profit'].plot(kind='bar', ax=axes[0,1], color='green', alpha=0.7)
axes[0,1].set_title('Total Profit by Category')
axes[0,1].set_xlabel('Category')
axes[0,1].set_ylabel('Profit')
axes[0,1].tick_params(axis='x', rotation=30)

# Profit Margin by Category
category_performance["Profit Margin"].plot(kind='bar', ax=axes[1,0], color='orange', alpha=0.7)
axes[1,0].set_title('Profit Margin by Category (%)')
axes[1,0].set_xlabel('Category')
axes[1,0].set_ylabel('Profit Margin')
axes[1,0].tick_params(axis='x', rotation=30)

# Order Count by Category
category_performance['Order Count'].plot(kind='bar', ax=axes[1,1], color='red', alpha=0.7)
axes[1,1].set_title('Order Count by Category')
axes[1,1].set_xlabel('Category')
axes[1,1].set_ylabel('Order Count')
axes[1,1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

#### ***Regional Performance*** ####

```python
print("********* Regional Performnace *******\n")

regional_performance = df.groupby('Region').agg({
    'Sales Amount':'sum',
    'Profit':'sum',
    'Order ID':'count'
}).rename(columns={'Order ID':'Order Count'})

regional_performance['Profit Margin'] = (regional_performance['Profit']) / (regional_performance['Sales Amount']) * 100
print(regional_performance,"\n\n")

#visualization
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
regional_performance['Sales Amount'].plot(kind='pie', autopct='%1.1f%%')
plt.title('Total Sales by Region')

plt.subplot(2,2,2)
regional_performance['Profit'].plot(kind='pie', autopct='%1.1f%%')
plt.title('Total Profit by Region')

plt.subplot(2,2,3)
sns.barplot(x=regional_performance.index, y='Order Count', data = regional_performance)
plt.title('Order Count by Region')
plt.xticks(rotation=30)

plt.subplot(2,2,4)
sns.barplot(x=regional_performance.index, y='Profit Margin', data = regional_performance)
plt.title('Profit Margin Region (%)')
plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

#### ***Time Series Analysis*** ####

```python
print("***** Time Series Analysis******\n")

#Extract time components
df['Order Month'] = df['Order Date'].dt.to_period('M')
df['Order Year'] = df['Order Date'].dt.year

#Monthy trends
monthly_trends = df.groupby('Order Month').agg({
    'Sales Amount':'sum',
    'Profit':'sum',
    'Order Date':'count'
}).rename(columns={'Order Date':'Order Count'})

print("Monthly Trends:\n")
print(monthly_trends,"\n\n")

plt.figure(figsize=(15, 10))

plt.subplot(3, 1, 1)
monthly_trends['Sales Amount'].plot(kind='line', color='blue', marker='o')
plt.title('Monthly Sales Trend')
plt.ylabel('Sales Amount')
plt.grid(True)

plt.subplot(3, 1, 2)
monthly_trends['Profit'].plot(kind='line', color='green', marker='o')
plt.title('Monthly Profit Trend')
plt.ylabel('Profit')
plt.grid(True)

plt.subplot(3, 1, 3)
monthly_trends['Order Count'].plot(kind='line', color='red', marker='o')
plt.title('Monthly Order Count')
plt.ylabel('Order Count')
plt.grid(True)

plt.tight_layout()
plt.show()
```

#### ***Advanced Analysis*** ####

```python
print("******* Correlation Analysis *********\n")

#select numerical columns for correlations
numerical_df = df[['Quantity','Unit Price','Discount (%)','Sales Amount','Cost Price','Profit']]

#Correlaion matrix
correlation_matrix = numerical_df.corr()
print("Correlation Matrix:")
print(correlation_matrix,"\n\n")

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, square=True, linewidths=0.5)
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

#### ***Customer Segment Analysis*** ####

```python
print("******* Customer Segment Analysis ******")

segment_analysis = df.groupby("Customer Segment").agg({
    'Sales Amount':['sum','mean'],
    'Profit':['sum','mean'],
    'Order ID':'count',
    'Discount (%)':'mean'
}).round(2)

segment_analysis.columns = ['Total Sales','Avg Sales','Total Profit','Avg Profit','Order Count','Avg Discount']
print(segment_analysis,"\n\n")

#Visualization

fig, axes = plt.subplots(3, 2, figsize=(15, 15))

segment_analysis['Total Sales'].plot(kind='bar', ax=axes[0,0], color='lightblue')
axes[0,0].set_title('Total Sales by Customer Segment')
axes[0,0].tick_params(axis='x', rotation=30)

segment_analysis['Avg Sales'].plot(kind='bar', ax=axes[0,1], color='orange')
axes[0,1].set_title('Average Sales by Customer Segment')
axes[0,1].tick_params(axis='x', rotation=30)

segment_analysis['Total Profit'].plot(kind='bar', ax=axes[1,0], color='lightgreen')
axes[1,0].set_title('Total Profit by Customer Segment')
axes[1,0].tick_params(axis='x', rotation=30)

segment_analysis['Avg Profit'].plot(kind='bar', ax=axes[1,1], color='red')
axes[1,1].set_title('Average Profit  by Customer Segment')
axes[1,1].tick_params(axis='x', rotation=30)

segment_analysis['Avg Discount'].plot(kind='bar', ax=axes[2,0], color='yellow')
axes[2,0].set_title('Average Discount by Customer Segment')
axes[2,0].tick_params(axis='x', rotation=30)

segment_analysis['Order Count'].plot(kind='bar', ax=axes[2,1], color='blue')
axes[2,1].set_title('Order Count by Customer Segment')
axes[2,1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```
